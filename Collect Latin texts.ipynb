```python
doc = fitz.open("decretum_gratiani.pdf")
print(doc)
```

```python
decretum_file = open("decretum.txt", "w")

for p, page in enumerate(doc):
    total_text = ""
    for word in page.get_text("words"):
        x0, y0, x1, y1, text, blocknumber, linenumber, wordnumber = word
        
        if p % 2 == 0 and (x0 > 140 and x1 < 430 and y0 > 80 and y1 < 570):
            total_text += text + " "
        elif p % 2 == 1 and (x0 > 185 and x1 < 475 and y0 > 80 and y1 < 465):
            total_text += text + " "
    decretum_file.write(total_text + "\n")
    #print(total_text)
    
decretum_file.close()    
```

```python
URL = "https://www.corpusthomisticum.org/iopera.html"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
f = open("corpus_thomisticum.txt", "w")

links = soup.find_all("a")
all_hrefs = []
for link in links:
    href = link.get("href")
    if href is None or href=="https://www.corpusthomisticum.org/" or "ealarcon" in href or "corpusthomisticum.org" not in href or link.string=="Nota editoris":
        continue
    print("Processing", href)
    all_hrefs.append(href)
    
    new_page = requests.get(href)
    new_soup = BeautifulSoup(new_page.content, "html.parser")
    for a_tag in new_soup.find_all('a'):
        a_tag.decompose()
    
    text = "\n".join(new_soup.text.split("\n")[14:-12])
    f.write(text)
f.close()
```

```python
base_url = "https://www.thelatinlibrary.com/"

already_visited = set()

def recursive_crawl(url):
    print("Processing", url)
    already_visited.add(url)
    page = requests.get(url)
    soup = BeautifulSoup(page.content, "html.parser")

    links = soup.find_all("a")
    all_text = ""
    
    for link in links:
        href = link.get("href")
        if href is None or href in ["index.html", "classics.html", "medieval.html", "/index.html", "/classics.html", "/medieval.html"] or "mailto" in href or "htm" not in href:
            continue
      
        new_url = base_url + href
        if new_url not in already_visited:
            all_text += recursive_crawl(new_url)
        
    for a_tag in links:
        a_tag.decompose()
    all_text += soup.text
    
    return all_text
    
all_text = recursive_crawl(base_url + "medieval.html")

# base_url = "https://www.thelatinlibrary.com/"
# page = requests.get(base_url + "medieval.html")

# soup = BeautifulSoup(page.content, "html.parser")
# #print(soup)

# links = soup.find_all("a")
# all_hrefs = []
# for link in links:
#     href = link.get("href")
#     if href is None or href in ["/index.html", "/classics.html", "/medieval.html"]:
#         continue
#     #if href is None or href=="https://www.corpusthomisticum.org/" or "ealarcon" in href or "corpusthomisticum.org" not in href or link.string=="Nota editoris":
#     #    continue
#     print("Processing", href, link.text)
#     new_page = requests.get(base_url + href)
#     new_soup = BeautifulSoup(new_page.content, "html.parser")
#     print(new_soup.text)
    
#     new_links = new_soup.find_all("a")
    
```
